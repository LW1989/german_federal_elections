```python
class ElectionModel:
    def __init__(self):
        self.ed_first_votes=None
        self.ed_second_votes=None
        self.struct_data=None
        self.features=None
        self.param_grid= {'n_estimators': [100, 200, 300, 400, 500],
                          'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3],
                          'max_depth': [2, 4, 6, 8, 10],
                          'colsample_bytree': [0.3, 0.5, 0.7, 0.9, 1],
                          'subsample': [0.2, 0.4, 0.6, 0.8, 1]
                         }
        self.model_dict={}
        
    def load_data(self,
                  path_to_first_votes,
                  path_to_second_votes,
                  path_to_struct_data,
                  ):
        
        first_votes=pd.read_csv(path_to_first_votes, index_col=[0])
        first_votes=first_votes.iloc[:,7:].fillna(0)
        first_votes['CDU/CSU_First_Votes_Final']=first_votes['CDU_First_Votes_Final']+first_votes['CSU_First_Votes_Final']
        first_votes.drop(columns=['CDU_First_Votes_Final','CSU_First_Votes_Final'], inplace=True, axis=0)
        self.ed_first_votes=first_votes
        
        second_votes=pd.read_csv(path_to_second_votes, index_col=[0])
        second_votes=second_votes.iloc[:,7:].fillna(0)
        second_votes['CDU/CSU_Second_Votes_Final']=second_votes['CDU_Second_Votes_Final']+second_votes['CSU_Second_Votes_Final']
        second_votes.drop(columns=['CDU_Second_Votes_Final','CSU_Second_Votes_Final'], inplace=True, axis=0)
        self.ed_second_votes=second_votes

        struct_data=pd.read_csv(path_to_struct_data, index_col=[0])
        self.struct_data=struct_data
        self.features=struct_data.columns
        
        
        
    def pre_process_data(self, feature_list):
        struct_data=struct_data.loc[:, feature_list]
        self.struct_data=struct_data
        
        
        
        
    #def train_model(self):
        
        
    #def predict(self):
    
    
        
```
